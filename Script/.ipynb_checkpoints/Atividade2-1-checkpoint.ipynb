## How To Isolate Trend, Seasonality And Noise From A Time Series
### BANCO DE DADOS - PLUVIOMETRIA ESTAÇÃO GOVERNADOR CELSO RAMOS

##### A time series can be thought of as being made of 4 components:
- Seasonal component
- Trend component
- Cyclical component
- Noise Componentnent.

```python
import pandas as pd
import numpy as np
import math
from matplotlib import pyplot as plt

# Carregando o arquivo modificado
caminho_arquivo_modificado = "C:\\Users\\edinei.acm\\ENS410064_OK\\ENS410064\\Dados\\Atividade2\\ChuvasPythonModificado.csv"
df = pd.read_csv(caminho_arquivo_modificado, parse_dates=['Mes'], dayfirst=True,  date_format='%m/%Y')
df.set_index('Mes', inplace=True)

# Plotando o conjunto de dados
fig = plt.figure()
fig.suptitle('Total de chuvas por mês')
df['TOTAL'].plot()
plt.show()

# Adicionando uma coluna com média móvel centrada 2 x 12 para capturar a tendência (6 meses antes, 6 meses depois
df['2 x 12 CMA (TREND)'] = df['TOTAL'].rolling(window=12, center=True).mean()

# Plotando a tendência
fig = plt.figure()
fig.suptitle('Componente de tendência das chuvas')
df['2 x 12 CMA (TREND)'].plot()
plt.show()

# Calculando a sazonalidade e o ruído
df['SEASONALITY AND NOISE'] = df['TOTAL']/df['2 x 12 CMA (TREND)']

# Plotando a sazonalidade e o ruído
fig = plt.figure()
fig.suptitle('Componentes de sazonalidade e ruído')
plt.ylim(0, 1.3)
df['SEASONALITY AND NOISE'].plot()
plt.show()

# Calculando o componente sazonal médio para cada mês
df['MONTH'] = df.index.month
average_seasonal_values = df.groupby('MONTH')['SEASONALITY AND NOISE'].mean().to_dict()
df['SEASONALITY'] = df['MONTH'].map(average_seasonal_values)

# Plotando o componente sazonal
fig = plt.figure()
fig.suptitle('Componente de sazonalidade')
plt.ylim(0, 1.3)
df['SEASONALITY'].plot()
plt.show()

# Calculando o ruído
df['NOISE'] = df['SEASONALITY AND NOISE']/df['SEASONALITY']

# Plotando o ruído
fig = plt.figure()
fig.suptitle('Componente de ruído')
plt.ylim(0, 1.3)
df['NOISE'].plot()
plt.show()

# Fazendo a decomposição sazonal usando a biblioteca statsmodels
from statsmodels.tsa.seasonal import seasonal_decompose
components = seasonal_decompose(df['TOTAL'], model='additive', period=12)
components.plot()
plt.show()

components.plot()
plt.show()
```
