```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
shape_path=r"D:\Mestrado\Dados Geoespaciais\Municípios\SC_municipios_2022.shp"
sc = gpd.read_file(shape_path)
fig,ax = plt.subplots()
sc.plot(ax=ax)
```

```python
import rasterio
import rioxarray as rioxr
from affine import Affine
```

```python
sc
```

```python
xaxim=sc[sc['NM_MUN']=='Xaxim']
xaxim.plot()
plt.show()
```

```python
#Raster utilizdo: suscetibilidade dos solos a erosão hídrica, fonte: Embrapa
path =r"D:\Mestrado\Dados Geoespaciais\Sensibilidade a Erosão\Brasil_suscetibilidade\Suscetibilidade.tif"
dataset = rasterio.open(path)
```

```python
dataRaster = rioxr.open_rasterio(path)
dataRaster.shape
```

```python
# Abrindo raster e recortando

from rasterio.mask import mask
# Caminho para o arquivo raster: path
# Caminho para o arquivo shapefile: shape_path

# Abre o arquivo raster
with rasterio.open(path) as src:
    # Recorta o raster usando as geometrias do shapefile
    clipped_data, _ = mask(src, xaxim.geometry, crop=True)

# Plota o resultado do raster recortado
plt.imshow(clipped_data[0], cmap='viridis')  # primeira banda
plt.colorbar()
plt.title("Suscetibilidade a erosão em Xaxim")
plt.show()
```

```python
# fazer média da suscetibilidade e inserir no shape
import numpy as np

# Calcula a média dos valores do raster
suscetibilidade_media = np.nanmean(clipped_data)

# Atribui a média a todas as entradas da coluna 'Suscetibilidade'
xaxim.loc[:,'SuscetibilidadeErosaoMedia'] = suscetibilidade_media
xaxim
```

```python
# fazer para todo o estado
from mpl_toolkits.axes_grid1 import make_axes_locatable

with rasterio.open(path) as src:
    # Crie uma nova coluna 'Suscetibilidade' no GeoDataFrame para armazenar as médias
    sc['Suscetibilidade'] = None

    # Itera pelas geometrias do GeoDataFrame (cada município)
    for index, row in sc.iterrows():
        # Recorta o raster usando a geometria do município atual
        clipped_data, _ = mask(src, [row.geometry], crop=True)

        # Calcula a média dos valores recortados
        suscetibilidade_media = np.nanmean(clipped_data)

        # Atribui a média à coluna 'Suscetibilidade' para o município atual
        sc.at[index, 'Suscetibilidade'] = suscetibilidade_media

# Plot do resultado com legenda de classes
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)

# Plote o GeoDataFrame com a legenda
plot = sc.plot(column='Suscetibilidade', cmap='viridis', cax=cax, ax=ax)
plot.set_title("Suscetibilidade Média por Município")

```
