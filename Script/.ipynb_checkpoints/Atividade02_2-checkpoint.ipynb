## The white noise model

### White noise detection using Python

```python
import pandas as pd
import numpy as np
import math
import statsmodels.graphics.tsaplots as tsa
from matplotlib import pyplot as plt

# Carregando o arquivo modificado
caminho_arquivo_modificado = "C:\\Users\\edinei.acm\\ENS410064_OK\\ENS410064\\Dados\\Atividade2\\ChuvasPythonModificado.csv"
df = pd.read_csv(caminho_arquivo_modificado, parse_dates=['Mes'], dayfirst=True,  date_format='%m/%Y')
df.set_index('Mes', inplace=True)

#df.head(10)

#correlacao
tsa.plot_acf(df['TOTAL'], lags=40, alpha=0.05, title='Auto-correlation coefficients for lags 1 through 40')
```

##### There is wave-like pattern in the auto-correlation plot that indicates that there could be some seasonality contained in the data. We can try to identify and isolate the seasonality by decomposing the time series into the trend, seasonality and noise components.

### The Chi-squared test for white noise detection#### 
The Chi-squared test is based on this powerful result in statistics: the sum of squares of k identical standard normal random variables is a Chi-squared distributed random variable with k degrees of freedom.

```python
#  Testing for white noise using the Ljung-Box test in Python
import statsmodels.stats.diagnostic as diag
 
diag.acorr_ljungbox(df['TOTAL'], lags=[40], boxpierce=True, model_df=0, period=None, return_df=None)
```

######  The value 759.0938398 is the value of the test statistic for the Ljung-Box test and 1.282463e-133 is its p-value as per the Chi-square(k=40)table. 
###### 
The valu728.91687948 is the test statistic of the Box-Pierce test an2.123791e-127.0 is its p-value as per the Chi-square(k=40) table 
###### 

As we can see, both p-values are less than 0.01 and so we can say with 99% confidence that moth rainfall lel level time series is not pure white noise.

```python
# take the first difference of the data, i.e. we create a new data set with Y = Y_i Y_(i-1) :

diff_Y_i = df['TOTAL'].diff()
#drop the NAN in the first row
diff_Y_i = diff_Y_i.dropna()
 
#Lets plot the diff-ed data set
diff_Y_i.plot()
plt.show()
```

```python
# Zoom para os ltimo s 2 anos
data_inicial = '2014-01-01'
data_final = '2016-12-01'
diff_Y_i.plot()
plt.xlim(data_inicial, data_final)
plt.show()
```

```python
# Novo Ljung Box test
diag.acorr_ljungbox(diff_Y_i, lags=[40], boxpierce=True)
```

##### This is easily enough to support the null hypothesis that the data (i.e. the differenced time series) is pure white noise.
#### The conclusion to be drawn from this exercise is that one should not fit anything except the White Noise model on this data.
